# Basics

## Exercise 1: Calculate Energy Consumption
Modify the values for power and time to calculate energy consumption for different devices, like a laptop (50W for 8 hours).

```python
# Calculate energy consumption in kWh
power = 60  # Power in watts (e.g., a light bulb)
time = 5  # Time in hours
energy_consumption = (power / 1000) * time  # Convert watts to kilowatts
print(f"Energy consumption: {energy_consumption} kWh")
```

## Exercise 2: Carbon Footprint Estimation


```python
# CO2 emissions per kWh (in kg, global average)
co2_per_kwh = 0.5
energy_used = 10  # Energy in kWh
carbon_footprint = energy_used * co2_per_kwh
print(f"Estimated CO2 emissions: {carbon_footprint} kg")
```

# Visualizing Energy Data

## Exercise 3: Visualize Appliance Consumption

```python
import matplotlib.pyplot as plt

# Example data
appliances = ['Fridge', 'TV', 'Laptop', 'AC', 'Heater']
consumption = [150, 100, 50, 2000, 1500]  # Power in watts

plt.bar(appliances, consumption)
plt.title("Energy Consumption of Appliances")
plt.xlabel("Appliances")
plt.ylabel("Power (W)")
plt.show()
```

## Exercise 4: Daily Energy Use Pie Chart
Display energy distribution for a day.

```python
import matplotlib.pyplot as plt

# Example data
activities = ['Heating', 'Cooling', 'Lighting', 'Electronics']
percentages = [40, 20, 25, 15]

plt.pie(percentages, labels=activities, autopct='%1.1f%%', startangle=140)
plt.title("Daily Energy Use Distribution")
plt.show()
```

# Predicting Energy Savings

## Exercise 5: Calculate Savings from LED Lighting
Change the input values for hours_per_day or electricity_cost to explore other scenarios.

```python
# Calculate energy and cost savings
old_bulb_power = 60  # Watts
led_power = 10  # Watts
hours_per_day = 5
days_per_year = 365
electricity_cost = 0.15  # Cost per kWh in USD

old_energy = (old_bulb_power / 1000) * hours_per_day * days_per_year
led_energy = (led_power / 1000) * hours_per_day * days_per_year
savings = (old_energy - led_energy) * electricity_cost

print(f"Annual savings by switching to LED: ${savings:.2f}")
```

### Display Geospatial Energy Data

### Energy Data Analysis with Pandas

```python
import pandas as pd

# Example data
data = {'Device': ['Fridge', 'TV', 'Laptop', 'AC'],
        'Old Power (W)': [150, 100, 50, 2000],
        'New Power (W)': [120, 80, 40, 1800]}

df = pd.DataFrame(data)
df['Savings (W)'] = df['Old Power (W)'] - df['New Power (W)']

print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Daten simulieren
hours = np.arange(24)
devices = {'Heater': 0.8, 'Fridge': 0.1, 'Lights': 0.2, 'Laptop': 0.05}  # kWh/h
total_consumption = sum([rate * hours for rate in devices.values()])

plt.stackplot(hours, *[rate * hours for rate in devices.values()], labels=devices.keys())
plt.legend(loc='upper left')
plt.title("Simulated Hourly Energy Consumption")
plt.xlabel("Hour of the Day")
plt.ylabel("Energy Consumption (kWh)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dummy-Daten erstellen
data = {'Date': pd.date_range(start='2023-01-01', periods=30),
        'Electricity Price (USD)': np.random.uniform(0.1, 0.3, 30)}
df = pd.DataFrame(data)

# Plotten der Daten
plt.plot(df['Date'], df['Electricity Price (USD)'])
plt.title("Electricity Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.xticks(rotation=45)
plt.show()
```
